## 코드 통합

```python
keyword = '박명수짤'
url = f"https://search.naver.com/search.naver?where=image&sm=tab_jum&query={keyword}"
driver = wb.Chrome()
driver.get(url)

srcList = []
# 1. 웹페이지 스크롤 내리기
for i in range(1) :
    print(i)
    driver.find_element(By.CSS_SELECTOR,'body').send_keys(Keys.END)
    time.sleep(1.5)
    
# 2. 이미지 요소 접근하기
    imgList = driver.find_elements(By.CSS_SELECTOR,'.tile_item._item > div > div > a > img')

# 3. 이미지 요소가 가진 src 속성값 접근하기
    for i in range(len(imgList)) :
        # h로 시작하는 값만 리스트에 저장하기!
        if imgList[i].get_attribute('src')[0] == 'h' :
            srcList.append(imgList[i].get_attribute('src'))

srcList
```

```python
# 4. 이미지주소로 특정 폴더에 파일 저장하기
from urllib.request import urlretrieve


for i in range(len(srcList)):
    urlretrieve(srcList[i],f'박명수짤/park{i+1}.jpg')
```

```python
keyword = '박명수짤'
if os.path.isdir(keyword) == True :
    print("이미 폴더가 있습니다.")
else :
    os.mkdir(keyword)
    print(f'{keyword}폴더를 생성하였습니다.')

# 이미지 주소로 특정 폴더에 파일 저장하기(위에서 리스트에 담아둔거)
for i in range(len(srcList)) :
    urlretrieve(srcList[i],f'{keyword}/cat{i+1}.jpg')
```
