### 멜론 음원데이터 수집 실습
1. 곡명과 가수 2가지 데이트 수집
2. 수집된 데이터를 엑셀(.xlsx)형식으로 저장 

```python
import requests as req
import pandas as pd  # csv 저장할때 필요하다고
from bs4 import BeautifulSoup as bs

h = {
"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"

}
# 헤더값 그냥 아까 썼던거 가져와도됨
# 브라우저의 언어 형식/어떤걸 가지고 왔냐? 그걸 보여주는게 user:Agent
# referer : 현재 페이지의 이전 페이지를 가리킨다. --> 네이버 웹툰 같은거 가져올때 사용을 했었다고 하심
# --> 참고 ! // https://iwantadmin.tistory.com/249
```

```python
melon_url = 'https://www.melon.com/chart/index.htm'
res = req.get(melon_url, headers = h)   # 데이터 변수에다 저장
# headers가 없다고 해서 406 오류가나옴 / headers추가하면 200으로 정상 출력됨
```

```python
res.text
```

```python
# 변환해준 친구를 가지고옴
html = bs(res.text , 'lxml')
```

```python
# '곡명' 수집하기
# 아이디가 없고 클래스를 가지고 온다고 하면/ ctrl+f 로 몇개 존재하는지 알아보고 떼어서 생각좀 해봐야해
#  그냥 다 가져와도 상관없음
musicList = html.select('.ellipsis.rank01 a')


# vocal = html.select('.ellipsis.rank02') # 이걸로 하면 두개씩 가져와짐(이유가 2시 40분 녹화에 있는데, 내가 제대로 못들음 다시 들어보기)
vocal = html.select('.checkEllipsis')
```

```python
for mu in musicList :
    print(mu.text.strip())
```

```python
for vo in vocal :
    print(vo.text.strip())
```

```python
# 1. 빈리스트 생성하기
# 2. 생성된 빈 리스트에 음원데이터/ 가수데이터 저장하기

mu = []
for i in musicList :
    mu.append(i.text.strip())
    
mu
```

```python
vo = []
for i in vocal :
    vo.append(i.text.strip())
vo
```

```python
mu = []
for i in range(len(musicList)) :
    mu.append(i.text.strip())
    vo.append(i.text.strip())
    
mu
```

```python
data_mu = pd.DataFrame(mu,vo)
data_mu
```

```python
# 3. 2개의 리스트를 활용해서 딕셔너리 생성하기
# 4. 생성된 딕셔너리로 데이터 프레임 만들기

melon_dic = { '곡명' : mu , '가수' : vo}
```

```python
meleon_df = pd.DataFrame(melon_dic)

meleon_df.index = meleon_df.index+1  # 인덱스 1부터 시작하기
meleon_df.head()  # 여기서 .head()는 위에서 다섯개/ .tail() 밑에서 다섯개
```

```python
# 5. 엑셀 파일로 저장하기(.xlsx)

meleon_df.to_excel('melon_top100.xlsx')   #현재위치에 저장을 하겟ㄱ다.



```
